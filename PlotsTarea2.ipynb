```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Heap

### Heap - \( v = 2^{10} \)

Se tienen los datos de los tiempos para v=2^10 en tiempos_heap_2e10.csv dentro de github, se importan:

```python
dfHeap10 = pd.read_csv('https://raw.githubusercontent.com/arriannee/logaritmos-tarea2/main/heap_csv/tiempos_heap_2e10.csv?token=GHSAT0AAAAAACTSEXYINPJFIOKZB2N3VKOSZTKTE4Q')
```

Se comprueba que los datos fueron importados correctamente

```python
dfHeap10.info()
```

Se muestra el gráfico y se realiza la regresión lineal

```python
model = LinearRegression()

aristas = dfHeap10['aristas'].values.reshape(-1,1)
tiempo = dfHeap10['tiempo']

model.fit(aristas,tiempo)
prediccion = model.predict(aristas)

plt.scatter(aristas, tiempo, label='e = 2^10', alpha=0.3)
plt.plot(aristas,prediccion, label='Regresión lineal')

plt.xlabel('Aristas')
plt.ylabel('Tiempo')
plt.title('Regresión Lineal Heap con v = 2^10 nodos')

coef = model.coef_[0]
intercept = model.intercept_

print(f"La pendiente de la recta es: {coef:.20f}")
print(f"Ecuación de la regresión lineal: tiempo = {intercept:.8f} + {coef:.20f} * aristas")
```
